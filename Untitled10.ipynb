```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(0)
df2 = pd.DataFrame(np.random.randn(10,4), columns=['A','B','C','D'])
```

```python
df2
```

```python
df2.style
```

```python
def style_negative(v, props=''):
    return props if v < 0 else None
```

```python
s2 = df2.style.applymap(style_negative, props='color:red;').applymap(lambda v: 'opacity: 50%;' if (v < 0.3) and (v > -0.3) else None)
s2
```
